```python
# Upload Data
PATH = "./data/origin/movielens.csv"
origin = pd.read_csv(PATH)
```

```python
# Positive feedback is limited to ratings of 3 or higher
pos_feedback = origin[origin["rating"]>=3]
```

```python
# Columns
kwargs = dict(
    df=pos_feedback,
    schema=schema,
    col_user='userId', 
    col_item='movieId', 
)

data = utils.preprocessor.main(**kwargs)
```

```python
# data Description
kwargs = dict(
    df=data,
    schema=schema,
    percentaile=0.9,
)

utils.description.main(**kwargs)
```
